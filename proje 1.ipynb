```python
mean_Western_Europe = Hapiness_report[Hapiness_report["Region"]=="Western Europe"]['Happiness_Score'].mean()
mean_North_America = Hapiness_report[Hapiness_report["Region"]=="North America"]['Happiness_Score'].mean()
mean_Australia_and_New_Zealand = Hapiness_report[Hapiness_report["Region"]=="Australia and New Zealand"]['Happiness_Score'].mean()
mean_Middle_East_and_Northern_Africa = Hapiness_report[Hapiness_report["Region"]=="Middle East and Northern Africa"]['Happiness_Score'].mean()
mean_Latin_America_and_Caribbean = Hapiness_report[Hapiness_report["Region"]=="Latin America and Caribbean"]['Happiness_Score'].mean()
mean_Southeastern_Asia = Hapiness_report[Hapiness_report["Region"]=="Southeastern Asia"]['Happiness_Score'].mean()
mean_Central_and_Eastern_Europe = Hapiness_report[Hapiness_report["Region"]=="Central and Eastern Europe"]['Happiness_Score'].mean()
mean_Eastern_Asia = Hapiness_report[Hapiness_report["Region"]=="Eastern Asia"]['Happiness_Score'].mean()
mean_Sub_Saharan_Africa = Hapiness_report[Hapiness_report["Region"]=="Sub-Saharan Africa"]['Happiness_Score'].mean()
mean_Southern_Asia = Hapiness_report[Hapiness_report["Region"]=="Southern Asia"]['Happiness_Score'].mean()
```
